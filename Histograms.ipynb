```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("firstasteroidData.txt")
ast_r = data[:,3]

# Set up plot window
fig, ax = plt.subplots(1,1,figsize=(10,6))

# Make histogram
counts, edges, patches = ax.hist(ast_r, bins=100,normed=False)
ax.set_title("Asteroid radius", fontsize=20)
ax.set_xlabel("Radius", fontsize=15)
ax.set_ylabel("Counts per bin", fontsize=15)
ax.tick_params(axis='x',labelsize=20)
ax.tick_params(axis='y',labelsize=20)
ax.set_xlim(2.,max(ast_r))

# Most prominent Kirkwood gaps
plt.axvline(x=2.06,color='red',linestyle='--')
plt.text(2.09, 1900,'4:1',size=15)
plt.axvline(x=2.5,color='red',linestyle='--')
plt.text(2.53, 1900,'3:1',size=15)
plt.axvline(x=2.82,color='red',linestyle='--')
plt.text(2.85, 1900,'5:2',size=15)
plt.axvline(x=2.95,color='red',linestyle='--')
plt.text(2.98, 1900,'7:3',size=15)
plt.axvline(x=3.27,color='red',linestyle='--')
plt.text(3.3, 1900,'2:1',size=15)

plt.show()
```
